# EDA(Explorative Data Analysis)

```python
# 1. Univarite Analysis

# Type of Employee
plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
data['employment_type'].value_counts().plot(kind = 'bar', color = sns.color_palette('tab10'))
plt.xlabel('Employment_type')
plt.ylabel('Employees')
plt.subplot(1,2,2)
data['employment_type'].value_counts().plot(kind = 'pie')
plt.show()
```

Bar plot: A bar plot or bar chart is a graph that represents the category of data with rectangle bars with length and heigths
    that is proportional to the values which they represent.

Pie chart: A Pie Chart is a circular statistical plot that can display only one series of data. 
    The area of the chart is the total percentage of the given data.
    The area of slices of the pie represents the percentage of th parts of the data

```python
# Level of Exprince
plt.figure(figsize =(15,5))
plt.subplot(1,2,1)
data['experience_level'].value_counts().plot(kind = 'bar', color = sns.color_palette('tab10'))
plt.subplot(1,2,2)
data['experience_level'].value_counts().plot(kind = 'pie', autopct = "%.2f")
plt.show()
```

```python
# Employees in every year

plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
data['work_year'].value_counts().plot(kind = 'bar', color = sns.color_palette('tab10'))
plt.subplot(1,2,2)
data['work_year'].value_counts().plot(kind = 'pie', autopct = "%.2f")
plt.show()
```

```python
# Remote Ratio
plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
data['remote_ratio'].value_counts().plot(kind = 'bar', color = sns.color_palette('tab10'))
plt.subplot(1,2,2)
data['remote_ratio'].value_counts().plot(kind = 'pie', autopct = '%.2f')
```

```python
# Company Size
plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
data['company_size'].value_counts().plot(kind = 'bar', color = sns.color_palette('tab10'))
plt.subplot(1,2,2)
data['company_size'].value_counts().plot(kind = 'pie', autopct = '%.2f')
```

```python
# Salary Currency
plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
data['salary_currency'].value_counts().plot(kind = 'bar', color = sns.color_palette('tab10'))
plt.subplot(1,2,2)
data['salary_currency'].value_counts().plot(kind = 'pie', autopct = '%.2f')
```

```python
plt.figure(figsize = (15,10))
# Exerience leve by year
plt.subplot(2,2,1)
sns.heatmap(pd.crosstab(data['work_year'], data['experience_level']))

# Employement_type by year
#FT employee is more in every year
plt.subplot(2,2,2)
sns.heatmap(pd.crosstab(data['work_year'], data['employment_type']))

# Remote ratio by year
plt.subplot(2,2,3)
sns.heatmap(pd.crosstab(data['work_year'], data['remote_ratio']))

# Size of companies by year
plt.subplot(2,2,4)
sns.heatmap(pd.crosstab(data['work_year'], data['company_size']))
```

Clustermap: Clustered heatmap are superior to ordinary heatmas as the heatmap cells are arranged in clusters.
    These clusters are further complemented by the dendograms which provide a summary view of the clusters including the inter cluster distances.

```python
sns.clustermap(pd.crosstab(data['work_year'], data['job_title']))
```

PairPlot: Plot pairwise relationship in a dataset. By default, this function will create a grid of Axes such that each numeric variable
    in data will be shared across the y-axes across a single row and the x-axes across a single column. The diagonal plots are treated
    differently: A univariate distribution plot is drawn to shw the marginal distribution of the data in each column.
        It is also possible to show a subset of variables or plot different variables on the rows and columns.
        This is a high-level interface for PaiGrid that is intended to make it easy to draw a few common styles.
        

```python
# Analyze entire dataset with pairplot
sns.pairplot(data,hue = 'company_size')
```

Analyze with Groupby

Pandas groupby: is used for grouping the data according to the categories and apply a function to the categories.
    It also helps to aggregate data efficiently.
    
Pandas dataframe.groupby() function is used to split the data into groups based on some criteria.
pandas object can be split on any of their axez.
The abstract definition of groupby is to provide a mapping of labels to group names.

```python
data.groupby(['company_size', 'experience_level', 'employment_type']).mean()
```

```python
plt.figure(figsize = (15,5))
data.groupby(['company_size', 'experience_level', 'employment_type']).mean()['salary'].plot(kind = 'bar', color = sns.color_palette('tab10'))
```
